## First analysis of MEDS dataset

We consider a collection of approximately 13000 temperature profiles previously quality controlled by expert examination, and attempt to reproduce those QC decisions at the per-profile level in software. A success rate in software of > 95% is demonstrated for flagging problematic profiles, with a false positive rate of < 0.1% (6 profiles out of ~8000). We conclude by demonstrating that these false positives are in fact profiles that should also be flagged as problematic, though at different levels than reported by expert QC.

### Data

 - Approximately 13000 XBT profiles from transcets IX01, PX11/IX22, PX34, PX30 and IX28 (Australian region) were considered.
 - Data was chosen for its thorough previous expert QC.
 - All data was encoded in MEDS format, and decoded using the `medspy.py` library in this repository.
 
### Methodology

 - While 12760 profiles were available in total, the present study considers only those profiles where every single level has received an expert QC assessment, leaving 12730 profiles to be considered.
 - Out of these 12730 profiles, only 87 profiles had no levels flagged as bad by expert QC. With such a statistically insignificant number of 'good' profiles, it would not be possible to make a meaningful argument for a low rate of false positives. As such, **we define a false positive as any QC decision in software that flags even a single level not flagged by expert QC**, even if the profile in question has other levels flagged as bad. In this sense, our false positive rate is understood as an upper bound at the per-profile level.
 - In order to give a meaningful estimate of our true positive rate (profiles flagged in software on a subset of levels also flagged by experts), **we exclude profiles flagged as wire breaks in software from the pool of 'bad' profiles to be flagged**. In other words, given that levels affected by a wire break are so easy to remove in software, we treat them as effectively 'good' profiles that don't require further examination, in order to avoid inflating our true positive rate with this well-known feature.
 
### Analysis

A simple framework to decode, sort and QC our data is presented in-line below; QC tests can be found in `medsqc.py`. Figures are produced for all false positives, with expert QC color-coded as blue for good levels, and red for levels of concern.

Besides CSIRO's wire break test, only very simple QC tests are performed here; one gradient test and three tests attempting to detect suspiciously constant temperatures below the mixing layer are performed.

```python
%matplotlib inline
import meds, medsqc, medshelpers, glob

files = glob.glob("data/*.MA")

reject_count = {
    'end_loose_constant': 0,
    'end_loose_loose_constant': 0,
    'end_strict_constant': 0,
    'simple_gradient': 0
}

good = 0
bad = 0
wire_breaks = 0
total = 0
bad_caught = 0

for f in files:
    data = meds.MEDSReader(f)
    for index, record in enumerate(data.records):
        p = meds.MEDSProfile(record)
        QC = p.t_QC()
        flagged = 0
        total += 1

        # don't even consider profiles that have even a single level that wasn't QCed:
        if 0 in QC:
            continue

        # flatten QC to simple bool flag / no flag
        QC = [x in [3,4] for x in QC]

        # wire break profiles will be considered otherwise 'good'
        if True in medsqc.csiro_wire_break(p):
            if sum(QC) == 0:
                print 'good profile got flagged by csiro_wire_break', f, index
            wire_break = True
            wire_breaks += 1
        else:
            wire_break = False

        # count total good / bad profiles
        if sum(QC)==0 or wire_break:
            good += 1
        else:
            bad += 1

        result = medsqc.simple_gradient(p)
        qc = medshelpers.process_result(QC, p, result, 'simple_gradient', f, index, wire_break)
        flagged = flagged or qc
        reject_count['simple_gradient'] += qc

        result = medsqc.end_loose_constant(p)
        qc = medshelpers.process_result(QC, p, result, 'end_loose_constant', f, index, wire_break)
        flagged = flagged or qc
        reject_count['end_loose_constant'] += qc

        result = medsqc.end_loose_loose_constant(p)
        qc = medshelpers.process_result(QC, p, result, 'end_loose_loose_constant', f, index, wire_break)
        flagged = flagged or qc
        reject_count['end_loose_loose_constant'] += qc
        
        result = medsqc.end_strict_constant(p)
        qc = medshelpers.process_result(QC, p, result, 'end_strict_constant', f, index, wire_break)
        flagged = flagged or qc
        reject_count['end_strict_constant'] += qc
        
        bad_caught += flagged

print 'total', total
print 'wire_breaks', wire_breaks
print 'good', good
print 'bad', bad
print 'bad_caught', bad_caught
print reject_count
```

### Discussion

As shown above, our four naive QC tests identify 4526 out of 4731 'bad' profiles (>95%), where 'bad' profiles are profiles with at least one level flagged in expert QC that was not a wire-break profile.

Of interest are the six false positives flagged by the simple gradient test. Four of them show sharp gradients that *were* flagged by expert QC, but only beginning part-way up the abrupt rise. Our simple gradient test would have flagged these at the first sharp gradient. A fifth (*Falkor2015*) shows similar behavior, where points are flagged only at the top of a 30 degree rise in 2 meters.

Less obvious above is exactly what automatic QC flagged that experts didn't in *st123041n2013*. We reproduce a detail of the plot below for closer inspection:

```python
import matplotlib
import matplotlib.pyplot as plt
import pylab

filename = 'data/st123041n2013.MA'
index = 26
data = meds.MEDSReader(filename)
p = meds.MEDSProfile(data.records[index])
QC = p.t_QC()
QC = [x in [3,4] for x in QC]

flags = [int(x) for x in QC]
title = filename[5:] + '-' + str(index)
fig = plt.scatter(p.z(),p.t(),c=flags,marker='o',lw=0)
plt.xlabel('Depth [m]')
plt.ylabel('Temperature [C]')
plt.title(title)
pylab.ylim([-10,40])
pylab.xlim([75,125])
plt.text(80,38, 'Lat: ' + str(p.latitude()))
plt.text(80,36, 'Long: ' + str(p.longitude()))
plt.text(80,34, 'Date: ' + str(p.year()) + '/' + str(p.month()) + '/' + str(p.day())    )
plt.show()
```

At this scale, the problem reveals itself to be the same as the rest; our simple gradient test would have flagged the jump of 10 degrees in less than a meter at the start of the scattered region above.
